```python
import random
# Generar una lista de números aleatorios enteros
random_list = [random.randint(1, 100) for _ in range(10)]
print(random_list)
print(sorted(random_list)) # sorted() devuelve una nueva lista ordenada sin modificar la original
random_list.sort()         # sort() ordena la lista original
print(random_list)

```

```python
# numero aleatorio entre 1 y 100 con 2 decimales
random_float = random.uniform(1, 100)
print(random_float)
print(f"{random_float:.2f}")
```

```python
# Generar un número aleatorio flotante entre 0 y 1
random_float = random.random()
print(random_float)
print("{:.3f}".format(random_float))
print(f"{random_float:.4f}")
print(round(random_float, 5))
```

```python
print(f'{1000000:,.2f}')   # separados por comas
print(f'{123:0<6d}')       # right padding with 0
print()
print(f'{123:0>6d}')       # left  padding with 0
print(str(5).zfill(3))
```
