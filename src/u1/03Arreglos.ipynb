```python

materia = "python"
print(materia[2]) # Seleccion de un carácter específico
for letra in materia:
    print(letra, end=" ")
print()

# Dividir la cadena en una lista de caracteres
caracteres = list(materia)
print(caracteres)
# usar split() para dividir la cadena en palabras
frase = "Python es un lenguaje de programación" 
palabras = frase.split()
print(palabras)


print(materia[::-1])  # Invertir la cadena
print(materia[::2])   # Seleccionar caracteres en posiciones pares
print("th" in materia)  # Verificar si "th" está en la cadena
print('1,2,3'.split(','))
print('1 2 3'.split(' '))

print(materia.find("c"))  # Encontrar la posición de "t"
try:
    print(materia.index("c"))
except ValueError:
    print("El carácter no se encuentra en la cadena")

print("Mi casa es tu casa".rfind("casa"))
```

```python
# Lists can be nested
my_list = [1, 'a', ["list", 64, [11, 22], False]]
print(my_list[2][2])
print(my_list[2][2][1])
```

```python
# Eliminar Espacios demas en una cadena
ciudad = "  San Luis         Potosí   "
estado = " ".join(ciudad.strip().split())
print("estado:", estado)

import re
resultado = re.sub(r"\s+", " ", ciudad).strip() # \s+ detecta uno o más espacios (o tabulaciones).
print(resultado)  # 👉 San Luis Potosí
print(resultado[:]) # all

```

```python
squares = [x ** 2 for x in range(10) if x**2 % 2 ==0] ## List comprehension
print(squares)

# 3 bildings, 15 floors each, 20 rooms on each floor
rooms = [[[False for r in range(20)] for f in range(15)] for b in range(3)]

```

```python
lista = [1, 2, 3, 4, 5]
y=list(map(lambda x:x+3,lista))  # lista es el valor de cada x
print("lista: ", y)
```

```python
mayores = list(filter(lambda x: x>2, lista)) 
print("Mayores:" , mayores)
```

```python
# Operador Ternario
# [código si se cumple] if [condición] else [código si no se cumple]
edad = 25
etapa = "adulto" if edad >= 20 else "menor"
print(etapa)

year = 2024
leap_year = "Sí" if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0) else "No"
print(f"¿El año {year} es bisiesto? {leap_year}")
```
