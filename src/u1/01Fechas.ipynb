```python
import datetime
print("Bienvenido al programa de Fechas.")

fecha = datetime.datetime.now()
print("La fecha y hora actual es:", fecha)

# dia = int(input("Ingrese el día (1-31): "))
```

```python
print("Year:", fecha.year)     
dia_semana = fecha.weekday() # 0 es Lunes, 1 es Martes, ..., 6 es Domingo
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
print("Día de la semana:", dias[dia_semana])
print("Dia de la semana:" , fecha.strftime("%A"))
```

```python
from datetime import datetime
fecha = datetime.strptime("13/03/2026", "%d/%m/%Y")
print("Fecha convertida:", fecha)
fecha = datetime(2020, 12, 25, 12, 30, 45)
print("Fecha y hora específica:", fecha)
```
